```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### 4H ETH TRADE ALGORITHM ###
df = pd.read_csv("tradeseth4H.csv", names = ["datetime", "action", "ETH", "return", "fee"],
                parse_dates = [0], index_col = 0)
```

```python
df.head()
```

```python
plt.figure(figsize = (16, 8))
plt.plot(df["return"])
plt.plot(df["ETH"])
plt.grid()
plt.axhline(1000, color = "red")
plt.xlabel("Date")
plt.ylabel("Dollars ($)")
plt.legend()
plt.plot()
```

```python
print("Current Equity ($):", df["return"][-1])
```

```python
print("Total Fees:", df.fee.sum())
```

```python
print("Returns (%):", (df["return"][-1] - df["return"][0]) / df["return"][0] * 100)
```

```python
print("Trade volume ($):", df["return"].sum())
```
